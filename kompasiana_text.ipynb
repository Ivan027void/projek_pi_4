```python
skipped_urls = []

for url in urls:
    # Kirim permintaan GET
    response = requests.get(url)

    # Parse konten dengan BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Temukan elemen dengan atribut itemprop="articleBody"
    article_body = soup.find(itemprop='articleBody')
    
    # Jika elemen tidak ditemukan, lewati url dan tambahkan ke list
    if article_body is None:
        skipped_urls.append(url)
        continue

    # Dapatkan teks dari setiap elemen <p> dan simpan ke dalam list
    paragraphs = [p.get_text() for p in article_body.find_all('p')]

    # Gabungkan list menjadi satu string, dengan setiap paragraf dipisahkan oleh baris baru
    text = '\n'.join(paragraphs)

    # Ekstrak judul halaman sebagai nama file (Anda mungkin perlu menyesuaikan ini tergantung pada struktur halaman web)
    title = soup.title.string

    # Buat versi singkat dari judul untuk digunakan sebagai nama file
    short_title = title[:50]  # Ambil 50 karakter pertama dari judul

    # Ganti karakter yang tidak valid dalam nama file
    filename = "".join(c for c in short_title if c.isalnum() or c.isspace()).rstrip() + ".txt"

    # Tentukan direktori tempat Anda ingin menyimpan file
    directory = "C:/Users/ahini/Downloads/projek_pi_4/scorpus_pi/"

    # Gabungkan direktori dan nama file untuk mendapatkan path lengkap ke file
    file_path = os.path.join(directory, filename)

    # Simpan teks ke file .txt
    with open(file_path, 'w') as f:
        f.write(text)
```

```python
import os
import re
import collections

# Define a directory where the txt files are stored
directory = "C:/Users/ahini/Downloads/projek_pi_4/scorpus_pi/"

# Create an empty dictionary to store the index
index = {}

# Loop through the files in the directory
for filename in os.listdir(directory):
    # Get the full path of the file
    filepath = os.path.join(directory, filename)
    # Open and read the file
    with open(filepath, "r") as file:
        text = file.read()
    # Split the text into words and lowercase them
    words = re.findall(r"\w+", text.lower())
    # Count the frequency of each word using a Counter object
    counter = collections.Counter(words)
    # Store the counter in the index with the filename as the key
    index[filename] = counter

# Define a search function that takes a query and returns a list of files
def search(query):
    # Split the query into words and lowercase them
    query_words = re.findall(r"\w+", query.lower())
    # Create an empty list to store the results
    results = []
    # Loop through the files in the index
    for filename, counter in index.items():
        # Check if all query words are in the counter
        if all(word in counter for word in query_words):
            # Add the filename to the results list
            results.append(filename)
    # Return the results list
    return results

# Define a rank function that takes a query and returns a sorted list of files by relevance
def rank(query):
    # Split the query into words and lowercase them
    query_words = re.findall(r"\w+", query.lower())
    # Create an empty list to store the results with scores
    results = []
    # Loop through the files in the index
    for filename, counter in index.items():
        # Check if all query words are in the counter
        if all(word in counter for word in query_words):
            # Calculate the score as the sum of frequencies of query words
            score = sum(counter[word] for word in query_words)
            # Add a tuple of filename and score to the results list
            results.append((filename, score))
    # Sort the results list by score in descending order
    results = sorted(results, key=lambda x: x[1], reverse=True)
    # Return only the filenames from the results list
    return [filename for filename, score in results]

# Test the search and rank functions with some queries
print(search("covid"))
print(rank("covid"))
print(search("vaccine"))
print(rank("vaccine"))
```

```python
# Import the modules
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
import os

# Define a directory where the txt files are stored
directory = "C:/Users/ahini/Downloads/projek_pi_4/scorpus_pi/"

# Create an empty list to store the documents
documents = []

# Loop through the files in the directory
for filename in os.listdir(directory):
    # Get the full path of the file
    filepath = os.path.join(directory, filename)
    # Open and read the file
    with open(filepath, "r") as file:
        text = file.read()
    # Append the text to the documents list
    documents.append(text)

# Create a TfidfVectorizer object with n-gram range of 1 to 2 and no maximum features
vectorizer = TfidfVectorizer(ngram_range=(1, 2))

# Fit and transform the documents into a TF-IDF matrix
tfidf_matrix = vectorizer.fit_transform(documents)

# Get the feature names
feature_names = vectorizer.get_feature_names_out()

# Print the feature names and their indices
print("Feature names and their indices:")
for i, name in enumerate(feature_names):
    print(i, name)

# Print the TF-IDF matrix as an array
print("\nTF-IDF matrix:")
print(tfidf_matrix.toarray())

# For each document, get the indices of the features with non-zero TF-IDF values
# and sort them by their values in descending order
print("\nRanking of features by document:")
for i, row in enumerate(tfidf_matrix):
    print(f"Document {i+1}:")
    # Get the non-zero indices and values
    nz_indices = row.nonzero()[1]
    nz_values = row[row.nonzero()].A1
    # Sort the indices by their values in descending order
    sorted_indices = nz_indices[np.argsort(nz_values)[::-1]]
    # Print the feature names and their values for each document
    for index in sorted_indices:
        print(f"{feature_names[index]}: {row[0, index]}")
```

```python
# Import cosine_similarity from sklearn.metrics.pairwise
from sklearn.metrics.pairwise import cosine_similarity

# Define a search function that takes a query and returns a list of documents
def search(query):
    # Transform the query into a TF-IDF vector using the same vectorizer
    query_vector = vectorizer.transform([query])
    # Create an empty list to store the results with scores
    results = []
    # Loop through the documents in the TF-IDF matrix
    for i, doc_vector in enumerate(tfidf_matrix):
        # Calculate the cosine similarity between the query vector and the document vector
        score = cosine_similarity(query_vector, doc_vector)
        # Add a tuple of document index and score to the results list
        results.append((i, score))
    # Sort the results list by score in descending order
    results = sorted(results, key=lambda x: x[1], reverse=True)
    # Return only the document indices from the results list
    return [index for index, score in results]
```
